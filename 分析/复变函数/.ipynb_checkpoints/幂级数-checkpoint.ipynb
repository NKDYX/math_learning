# 幂级数
所谓幂级数是指形如:
$\sum_{n=0}^{\infty}a_n(z-z_0)^n$的级数,自然，我们可以通过平移将级数化简成$\sum_{n=0}^{\infty}a_nz^n$进行研究

## 收敛半径
幂级数的重要性质是其有一个收敛圆，圆的内部都收敛，圆的外部都发散，圆上则要视情况判断。  
收敛半径恰好是{$\sqrt[n]{|a_n|}$}的上极限(数列未必存在极限，但是一定存在上极限和下极限，即是所有广义收敛(包括正负无穷)的子列的极限的上确界和下确界，因为实数集的子集必有上下确界是由实数戴德金分割的定义决定的,并且广义收敛使得这个极限的集合不会出现空集并出现上确界小于下确界的特殊情况)的倒数。上极限还有明确的等式表达{$a_n$}的上极限是$\lim\limits_{n\rightarrow\infty}{\underset{k\geq n}{sup}a_k}$  
我们分三种情况讨论:  
1. R=0,级数在z=0显然收敛，当$z\neq0$时，{$a_n$}必有递增子列。那么必有子列${\sqrt[n_{\,k}\,]{|a_{n_k}|}}$使得其中每一项都大于$\frac{1}{|z|}$，那么对应的$|a_{n_k}z^{n_k}|>1$,单考虑{$a_{n_k}z^{n_k}$}都不收敛，必然发散。  
2. R=$\infty$，z=0依旧收敛，对于复平面上其它点$z\neq 0$，由于上极限趋向于0，那么去除开头有限项后必然有子列${\sqrt[n]{|a_{n}|}}$使得其中每一项都小于$\frac{1}{2|z|}$，从而对应的$|a_{n}z^{n}|<\frac{1}{2^{n}}$,级数绝对收敛，自然收敛。  
3. $0<R<\infty$,在收敛圆内的z离原点的距离严格小于R,总能有$|z|<\rho<R$，从而去除开头有限项后有子列${\sqrt[n]{|a_n|}}$每一项都小于$\frac{1}{\rho}>\frac{1}{R}$，从而对应的$|a_nz^n|<(\frac{|z|}{\rho})^{n}$,绝对收敛必然收敛。在收敛圆外的z离原点的距离严格大于R,总能有$|z|>r>R$,从而有子列${\sqrt[n_{\,k}\,]{|a_{n_k}|}}$每一项都大于$\frac{1}{r}<\frac{1}{R}$，从而对应的$|a_{n_k}z^{n_k}|>(\frac{|z|}{r})^{n_k}>1$,单考虑{$a_{n_k}z^{n_k}$}都不收敛，必然发散。    

## 阿贝尔定理
$\quad$对于幂级数，逐点收敛的性质已经由收敛半径较好的描述了，然而一致收敛(内闭一致收敛)才是函数项级数保持连续性,逐项可积性(逐项可导性)的关键.所谓一致收敛，就是说函数项级数在定义域中每一点收敛的速度相同，这就要求，对于任意一个固定的误差$\epsilon$,我们有一个统一的N，使得定义域上每一点函数项级数的值和函数项级数部分和之间的误差不超过$\epsilon$。至于内闭一致收敛，这是一致收敛的必要条件，很多时候，不能一致收敛仅仅是因为定义域本身的性质不够好，如果函数项级数在定义域中每一个紧子集上一致收敛就称为内闭一致收敛(在欧氏空间中紧集等价于有界闭集很多时候我们都是在讨论一个有界集所以这里干脆就定义为在紧子集上一致收敛)。  
$\quad$阿贝尔定理证明的关键在于使用魏尔斯特拉斯收敛法，所谓魏尔斯特拉斯收敛法，就是可以对函数项级数的每一项构造一个关于复平面(实函数的x轴)对称的层状(带状)界限，而且层(带)宽组成的数列是收敛的，那么显然函数项级数是一致收敛的，因为下标越大，对应项在整个定义域上的最大波动越小，直到趋向于0。至于阿贝尔定理的证明。  
$\quad$阿贝尔定理是说，如果一个幂级数在非零点z收敛，那么以|z|为收敛半径的收敛圆内部(不包含边界)该幂级数内闭一致收敛。  
$\quad$证明本身是显然的，因为幂级数在z处收敛，那么$|a_nz^n|<M$是有界的，假如$z_0$是收敛圆内紧子集上的一点,由于紧子集上函数的最值可以取到，那么不妨设紧子集上离原点最远的距离是r，幂级数每一项自然就有魏尔斯特拉斯收敛法中要求的层状界限$|a_nz_0^n|=|a_nz^n\frac{z_0^n}{z^n}|\leq M\frac{|z_0|^n}{|z|^n}\leq M\frac{r^n}{|z|^n}$,而且这个界限的厚度是趋向于0的。  
## (内闭)一致收敛决定的复级数性质
1. 函数项级数$\sum_{n=1}^{\infty}f_n(z)$在任意点集E上一致收敛，那么如果每一项$f_n(z)$在E上连续，那么$f(z)$在E上连续。  
证明是非常简单的，由于不涉及到复变函数特殊的求导和数学分析中完全一致。即用三个因素的和构造的趋于0的上界控制$f(z)$在$z_0$的增长，这三个因素分别是$z_0$点部分和和级数的误差，z点部分和和级数的误差，以及部分和在$z_0$点和z点之间的误差。这三个因素当然都趋于0，前两个是由一致收敛决定的后一个是$f_n(z)$的连续性和连续性关于有限加法不变决定的。  
2.函数项级数$\sum_{n=1}^{\infty}f_n(z)$在可求长曲线$\gamma$上一致收敛到$f(z)$，那么如果每一项$f_n(z)$在E上连续，那么$\int_{\gamma}f(z)dz=\sum_{n=1}^{\infty}\int_{\gamma}f_n(z)dz$。  
证明同样和数学分析中完全一致。如果部分和以及函数项级数之间的误差是$\epsilon$那么$\epsilon |\gamma|$是部分积分和的一个明显上界。  
3.D是C中的域，如果$f_n(z)$在D上处处解析而且$\sum_{n=1}^{\infty}f_n(z)$在D上内闭一致收敛,那么函数项级数$f(z)=\sum_{n=1}^{\infty}f_n(z)$在D上处处解析,而且$\sum_{n=1}^{\infty}f_n^{(k)}(z)$在D上内闭一致收敛到$f^{(k)}(z)$。  
这里的结论和证明由于涉及到导数和数学分析是完全不同的。首先$f(z)$在D上处处解析可以通过1和2证明，因为在D中$f_n(z)$处处解析，那么每个可求长闭曲线上$f_n(z)$的积分都是0，由于解析必连续，相应级数$f(z)$对应的积分也是0而且$f(z)$保持连续性，那么连续函数$f(z)$在D上积分是路径无关的，故而存在函数$F(z)=\int_{z_0}^zf(z)dz$,不难证明该函数是f在D上的原函数，该原函数也是解析函数，从而f作为解析函数的导数也是解析函数。至于内闭一致收敛，只需要用柯西积分公式章节中提到的柯西不等式对于高阶导数求得一个上界即可证明，这里需要注意的是上界中有关距离的常数是在紧集上关于该点被包含的开球的半径变化的(之所以每个点有开球包含，因为域是非空连通开集)，但是紧集上的函数能够取到最值。
